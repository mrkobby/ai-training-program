```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
```

```python
url = 'http://dvla.gov.gh/tables.php'
```

```python
html = urlopen(url)
soup = BeautifulSoup(html,'lxml')
```

```python
all_tr = soup.find_all('tr')

list_tr = []

for td in all_tr:
    row_cells = td.find_all('td')
    str_cells = str(row_cells)
    str_cells = BeautifulSoup(str_cells,'lxml').get_text()
    list_tr.append(str_cells)
    
list_tr
```

```python
list_th = []

th_cells = soup.find_all('th')
str_th_cells = str(th_cells)
str_th_cells = BeautifulSoup(str_th_cells,'lxml').get_text()
list_th.append(str_th_cells)

list_th
```

```python
import pandas as pd

tr_df = pd.DataFrame(list_tr)
th_df = pd.DataFrame(list_th)
```

```python
tr_df = tr_df[0].str.split(',',expand=True)
tr_df
```

```python
th_df = th_df[0].str.split(',',expand=True)
th_df
```

```python
th_df[0] = th_df[0].str.strip('[')
th_df[4] = th_df[4].str.strip(']')
th_df
```

```python
tr_df[0] = tr_df[0].str.strip('[')
tr_df[4] = tr_df[4].str.strip(']')
tr_df
```

```python
tb_frames = [tr_df,th_df]
df = pd.concat(tb_frames)

df = df.rename(columns=th_df.iloc[0])
df
```

```python
df = df.drop(df.index[0])
df
```
