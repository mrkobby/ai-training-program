```python
#-- PRINTING ANOTHER INSTANCE OF STOCKS.CSV
dw = pd.read_csv('stocks.csv')
dw
```

```python
#-- 1ST GETTING THE INDEX AND SYMBOL OF MAX CLOSE COLUMN VALUE
dw.loc[dw['Close'] == dw['Close'].max()]['Symbol']
```

```python
#-- 2ND GETTING THE INDEX AND SYMBOL OF MAX CLOSE COLUMN VALUE
close_max = dw['Close'].max()
dw.loc[dw['Close'] == close_max,['Symbol']]
```
