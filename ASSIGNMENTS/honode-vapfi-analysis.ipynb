```python
#-- CONVERT CUURENT DATE FORMAT INTO A READABLE FORMAT - *RUN CODE ONCE
#-- FIRST IMPORT THE DATATIME LIBRARY
from datetime import datetime

#-- PUT ALL DATES IN THE DATAFRAME INOT A LIST
listdates = df['DateOfBirth'].tolist()

#-- RUN A LOOP THAT WILL CONVERT EACH DATE IN THE 'listdates' LIST INTO A READABLE FORMAT
for steps in listdates:
    dateofbirth_object = datetime.strptime(steps, '%Y-%m-%d').date()
    str_dtm = dateofbirth_object.strftime('%B %d, %Y')
    df.loc[df['DateOfBirth'] == steps,'DateOfBirth'] = str_dtm
```

```python
#-- PLOTTING THE PIE CHART GRAPH
#-- CREATE LIST OF LOCATIONS NEEDED ON THE GRAPH
location_list = ['HO','SOGAKOPE','AGBOZUME']

#-- GET NUMBER OF APPLICANTS FROM THOSE LOCATIONS IN THE LIST
ho = df.loc[df['Location'] == 'HO'].shape[0]
sogakope = df.loc[df['Location'] == 'SOGAKOPE'].shape[0]
agbozume = df.loc[df['Location'] == 'AGBOZUME'].shape[0]

#-- PUT NUMBERS OF APPLICANT LOCATIONS IN A LIST
applicant_location_count = [ho,sogakope,agbozume]

#-- PLOT CHART
plt.pie(applicant_location_count,labels=location_list,autopct='%1.2f%%', radius=1, shadow=True)
plt.title('Pie Chart of Locations and pecentage of Applicants')
```
