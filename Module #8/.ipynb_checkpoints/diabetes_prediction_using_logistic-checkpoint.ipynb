```python
import pyforest
from joblib import dump, load
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix
```

```python
data = pd.read_csv('diabetes.csv')
data.head()
```

```python
data['Glucose'] = data['Glucose'].astype(float)
data['Pregnancies'] = data['Pregnancies'].astype(float)
data['BloodPressure'] = data['BloodPressure'].astype(float)
data['SkinThickness'] = data['SkinThickness'].astype(float)
data['Insulin'] = data['Insulin'].astype(float)
data['Age'] = data['Age'].astype(float)
```

```python
plt.figure(figsize=(10,6))
plt.scatter(data['Glucose'], data['Age'])
plt.xlabel('Glucose')
plt.ylabel('Age')
plt.show()
```

```python
sns.lmplot('Glucose','Age',data=data)
```

```python
sns.heatmap(data.corr(), annot=True)
```

```python
features = data[['Pregnancies', 'Glucose' , 'BMI', 'DiabetesPedigreeFunction', 'Age', 'Insulin']]
label = data[['Outcome']]
```

```python
model = LogisticRegression()
```

```python
features_train, features_test, label_train, label_test = train_test_split(features, label, test_size=0.1)
```

```python
model.fit(features_train,label_train)
```

```python
print(classification_report(label_test, model.predict(features_test)))
```

```python
dump(model, 'diabetes_predict.joblib')
```

```python
print(confusion_matrix(label_test, model.predict(features_test)))
```
