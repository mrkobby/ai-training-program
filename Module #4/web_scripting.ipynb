```python
from urllib.request import urlopen
from bs4 import BeautifulSoup as soup
```

```python
url = 'https://www.hubertiming.com/results/2017GPTR?search=&DivisionFilter=F+26-35'
```

```python
html = urlopen(url)
soup = BeautifulSoup(html,'lxml')
```

```python
all_tr = soup.find_all('tr')
# print(all_tr[:10])
```

```python
#  for row in all_tr:
#     all_tr = row.find_all('td')
#     print(all_tr[:10])
```

```python
list_rows = []
for row in all_tr:
    cell = row.find_all('td')
    str_cell = str(cell)
    str_cell = BeautifulSoup(str_cell,'lxml').get_text()
    list_rows.append(str_cell)

list_rows
    
#type(cell)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(list_rows)
df.head(10)
```

```python
df1 = df[0].str.split(',',expand=True)
df1.head(10)
```

```python
df1[0] = df1[0].str.strip('[')
df1.head(10)
```

```python
df1[12] = df1[12].str.strip(']')
df1.head(10)
```

```python
list_th = []

cell1 = soup.find_all('th')
str_cell1 = str(cell1)
str_cell1 = BeautifulSoup(str_cell1,'lxml').get_text()
list_th.append(str_cell1)

list_th
```

```python
df2 = pd.DataFrame(list_th)
df2 = df2[0].str.split(',',expand=True)
#df2
```

```python
df2[0] = df2[0].str.strip('[')
df2[12] = df2[12].str.strip(']')
df2
```

```python
frames = [df2,df1]
df3 = pd.concat(frames)
df3.head(10)
```

```python
df3 = df3.rename(columns=df3.iloc[0])
df3.head(10)
```

```python
df3 = df3.drop(df3.index[0:4])
```
